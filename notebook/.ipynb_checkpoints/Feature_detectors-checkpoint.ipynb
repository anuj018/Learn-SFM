## Experimenting with the SIFT Feature Detector and Descriptor

```python
import numpy as np
import cv2 as cv
img = cv.imread('./../Feature_rich_image.jpg')
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create()
```

```python
sift.setContrastThreshold(0.1)
sift.setEdgeThreshold(15)
sift.setNOctaveLayers(2)
sift.setSigma(2)
```

```python
keypoints, descriptors = sift.detectAndCompute(gray, None)
img=cv.drawKeypoints(gray,keypoints,img)
cv.imwrite('sift_keypoints_for_Feature_rich_image.jpg',img)
```
